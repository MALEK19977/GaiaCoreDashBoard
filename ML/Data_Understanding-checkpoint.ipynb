```python
# Import libraries
import pandas as pd
import os
import json
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import seaborn as sns

# Define data folder and files
data_folder = r"C:\Users\21695\OneDrive\Desktop\BI Project pi"
target_files = {
    "json_file": "Climate_Environmental_Data.json",
    "excel_file": "Equipments.xlsx",
    "text_file": "Materials.txt",
    "xml_file": "Notifications.xml",
    "csv_file": "Suppliers.csv",
    "bak_file": "SAP.bak"
}

# Results container
data_overview = {}

def analyze_dataframe(df, name):
    """Perform data analysis on a given DataFrame"""
    print(f"\n===== {name} Overview =====")
    # Check basic info
    print(df.info())
    # Show first few rows
    print("\nFirst 5 Rows:")
    print(df.head())
    # Check missing values
    print("\nMissing Values:")
    print(df.isnull().sum())
    # Generate statistical summary for numeric columns
    print("\nDescriptive Statistics (Numeric):")
    print(df.describe())
    # Visualize missing data
    sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
    plt.title(f"Missing Values Heatmap for {name}")
    plt.show()

# a. JSON File: Climate_Environmental_Data.json
print("\n=== Processing JSON File ===")
json_path = os.path.join(data_folder, target_files["json_file"])
with open(json_path, 'r') as f:
    json_data = json.load(f)

climate_df = pd.json_normalize(json_data)
analyze_dataframe(climate_df, "Climate Environmental Data (JSON)")
data_overview["JSON"] = climate_df

# b. Excel File: Equipments.xlsx
print("\n=== Processing Excel File ===")
excel_path = os.path.join(data_folder, target_files["excel_file"])
equipment_df = pd.read_excel(excel_path)
analyze_dataframe(equipment_df, "Equipments Data (Excel)")
data_overview["Excel"] = equipment_df

# c. Text File: Materials.txt
print("\n=== Processing Text File ===")
txt_path = os.path.join(data_folder, target_files["text_file"])
materials_df = pd.read_csv(txt_path, sep="\t", header=None)  # Update separator and header as per file structure
analyze_dataframe(materials_df, "Materials Data (Text)")
data_overview["Text"] = materials_df

# d. XML File: Notifications.xml
print("\n=== Processing XML File ===")
xml_path = os.path.join(data_folder, target_files["xml_file"])
tree = ET.parse(xml_path)
root = tree.getroot()

# Convert XML to DataFrame
xml_data = []
for child in root:
    row = {elem.tag: elem.text for elem in child}
    xml_data.append(row)

notifications_df = pd.DataFrame(xml_data)
analyze_dataframe(notifications_df, "Notifications Data (XML)")
data_overview["XML"] = notifications_df

# e. CSV File: Suppliers.csv
print("\n=== Processing CSV File ===")
csv_path = os.path.join(data_folder, target_files["csv_file"])
suppliers_df = pd.read_csv(csv_path)
analyze_dataframe(suppliers_df, "Suppliers Data (CSV)")
data_overview["CSV"] = suppliers_df

# f. SAP.bak File: Requires manual handling as explained
print("\n=== Handling SAP.bak File ===")
print("For the .bak file, restore it using SQL Server Management Studio (SSMS), then export the tables to .csv or .xlsx for analysis.")

# Summary for all analyzed files
print("\n=== Summary of All Analyzed Files ===")
for file_type, df in data_overview.items():
    print(f"\n--- {file_type} Data Summary ---")
    print(f"Shape: {df.shape}")
    print(f"Missing Values:\n{df.isnull().sum()}\n")
    print(df.head())
```
